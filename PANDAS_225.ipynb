### Convert the dictionary into DataFrame

```python
rankings_pd = pd.DataFrame(rankings) # converting the dictionary into dataframe
```

```python
rankings_pd.rename(columns = {'test':'TEST'}, inplace = True)
```

```python
rankings_pd.rename(columns = {'odi':'ODI'}, inplace = True)
```

```python
rankings_pd.rename(columns = {'t20':'T20'}, inplace = True)
```
